```python
import pandas as pd 
import numpy as np 
import os
import matplotlib.pyplot as plt 
import cv2
```

```python
path_train = 'dataset/train/'
path_test = 'dataset/test/'
```

```python
dim = 28
labels = ['pixel'+str(n) for n in np.arange(28*28)]
pixels_all = []
```

```python
def generate_csv(path):
    imgs = []
    targets = []
    
    for digit in range(10):
        digit_folder = path + str(digit) + '/'
        digit_folder_imgs = os.listdir(digit_folder)
        for digit_file in digit_folder_imgs:            
            img = cv2.imread(digit_folder + digit_file)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            #print(len(img.flatten()))
            #print(len(img))
            #return 1 
            imgs.append(img.flatten())
            targets.append(digit)
    
    #print(len(imgs))
    #for img in imgs:
    #    pixels_img = []
    #    for pixel in img.flatten():
    #        pixels_img.append(pixel)
    #    
    #    pixels_all.append(pixels_img)
    
    #print(len(pixels_all))
    df = pd.DataFrame(data=imgs)
    df.columns = labels
    #print(df.shape)
    #print(len(targets))
    df /= 255
    df['target'] = targets
    df = df[np.append('target', df.columns[:-1])]
    return df
```

```python
train = generate_csv(path_train)
train
```

```python
test = generate_csv(path_test)
test
```

```python
train.to_csv('dataset/train.csv')
test.to_csv('dataset/test.csv')
```
